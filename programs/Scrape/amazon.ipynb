```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.amazon.in/s?k=mobiles&i=electronics&crid=3GUH7BWS28ZK&sprefix=mobile%2Celectronics%2C206&ref=nb_sb_noss_1"
```

```python
#headers for useragent

HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36','Accept-Language':'en-US, en;q=0.5'})
```

```python
# HTTP Request
webpage= requests.get(URL, headers=HEADERS)
webpage
```

```python
type(webpage.content)
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
soup
```

```python
#Fetch links as List of Tag Objects
links = soup.find_all("a",attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})

links
```

```python
link = links[1].get('href')
```

```python
product_list = "https://amazon.in" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
```

```python
new_soup
```

```python
# productTitle
new_soup.find("span", attrs={'id':'productTitle'})
```

```python
# Price
new_soup.find('span', attrs={'class':'a-price aok-align-center reinventPricePriceToPayMargin priceToPay'}).find('span',attrs={'class':'a-offscreen'}).text
```

```python
#Ratings
new_soup.find("span", attrs={'id':'acrCustomerReviewText'}).text.strip()
```

```python
#Review
new_soup.find("i", attrs={'class':'a-icon a-icon-star a-star-4-5 cm-cr-review-stars-spacing-big'}).text.strip()
```
